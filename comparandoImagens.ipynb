```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from colorama import init
from termcolor import colored
def mostraImagem(img):
    if img is not None:
        plt.figure(figsize=(20,10))
        plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
    else:
        print('img Ã© nula')
def mediaImagem(img):
    # find average per row, assuming image is already in the RGB format.
    # np.average() takes in an axis argument which finds the average across that axis.
    average_color_per_row = np.average(img, axis=0)

    # find average across average per row
    average_color = np.average(average_color_per_row, axis=0)

    # convert back to uint8
    average_color = np.uint8(average_color)
    #print(average_color)
    return average_color

def avaliarArquivos():
    import glob

    lista = [f for f in glob.glob('img/*.jpg')]

    for i,item in enumerate(lista):
        if (len(lista) > i+2 ):
            original = cv2.imread(lista[i])
            duplicate = cv2.imread(lista[i+1])
            
            if original.shape == duplicate.shape:
                print("--------------------------------------------------------------")
                print(colored("as imagens {} e {} tem o mesmo camanho e canal".format(lista[i],lista[i+1]),'blue'))
                diferenca = cv2.subtract(original, duplicate)
                b, g, r = cv2.split(diferenca)
                
                verificacao01 = mediaImagem(original - duplicate)
                verificacao02 = mediaImagem(duplicate - original)
                
                if np.array_equal(verificacao01,verificacao02):
                    print(colored('imagens identicas','green'))
                else:
                    print(colored('imagens diferentes','red'))
                
                
                
```
